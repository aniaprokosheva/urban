## Проект: Объекты из OSM

**Цель проекта:** выделить на карте объекты г. Киров используя данные из OSM.

**Библиотеки:** keplergl, osmnx

```python
# для отображения Kepler в colab
from keplergl import KeplerGl
import osmnx as ox

from google.colab import output
output.enable_custom_widget_manager()
```

```python
#город
CITY_NAME = 'Киров'

city = ox.geocode_to_gdf(CITY_NAME)
city.explore(tiles="CartoDB positron")
```

```python
# район города
REGION_NAME = 'Нововятский район, Киров'

region = ox.geocode_to_gdf(REGION_NAME)
region.explore(tiles="CartoDB positron")
```

```python
# жилые дома по тегу 'building': 'apartments' для района
tags = {'building': 'apartments'}
buildings = ox.geometries_from_polygon(region.unary_union, tags)
# добавим индексы
buildings.reset_index(inplace=True)
buildings.head()
```

```python
#дома на карте
buildings.explore(tiles="CartoDB positron")
```

```python
# автобусные остановки по тегу 'highway': 'bus_stop' для района
tags = {'highway': 'bus_stop'}
bus_stop = ox.geometries_from_polygon(region.unary_union, tags)
# индексы
bus_stop.reset_index(inplace=True)
bus_stop.head()
```

```python
# на карте
bus_stop.explore(tiles="CartoDB positron")
```

```python
# на карте Kepler границы города и Нововятского района, жилые дома и автобусные остановки
map = KeplerGl(height=500)
map.add_data(data=buildings.copy(), name='Жилые дома')
map.add_data(data=bus_stop.copy(), name='Автобусные остановки')
map.add_data(data=region.copy(), name='Нововятский район')
map.add_data(data=city.copy(), name='Киров')
map
```

**ВЫВОД:** Визуализировали жилые дома типа apartments и автобусные остановки в Нововятском районе города Кирова.
