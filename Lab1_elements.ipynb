```python
from keplergl import KeplerGl
import osmnx as ox
```

```python
CITY_NAME = 'Санкт-Петербург'

city = ox.geocode_to_gdf(CITY_NAME) # информация из OSM превращается в геодф
city.explore(tiles="CartoDB positron")  # интерактивная карта
```

```python
tags = {'admin_level': '5'}  # ранг в системе тегов OSM у районов города
border = ox.geometries_from_polygon(city.unary_union, tags)   # информация из OSM превращается в геодф 
# unary_union возвращает все геометрии из гдф и объединяет их
border.explore(tiles = 'CartoDB positron')  # интерактивная карта
```

```python
# выделим нужные районы
area_petr_name = 'Петроградский район, Санкт-Петербург'
area_petr = ox.geocode_to_gdf(area_petr_name)
area_petr.explore(tiles = 'CartoDB positron')  # интерактивная карта
```

```python
area_vas_name = 'Василеостровский район, Санкт-Петербург'
area_vas = ox.geocode_to_gdf(area_vas_name)
area_vas.explore(tiles = 'CartoDB positron')  # интерактивная карта
```

```python
tags = {'building': True} # здания
buildings = ox.geometries_from_polygon(area.unary_union, tags)  
buildings.reset_index(inplace = True)
buildings = buildings[buildings.element_type != 'node'] # удаляю здания, обозначенные как точки
```

```python
tags = {'leisure': 'park'}  # leisure=park
parks = ox.geometries_from_polygon(area_vas.unary_union, tags)
parks.reset_index(inplace=True)
parks = parks[parks.element_type != 'node'] # удалим парки, обозначенные как точки
```

```python
map = KeplerGl(height = 500)
map.add_data(data=buildings.copy(), name='buildings')
map.add_data(data = parks.copy(), name = 'parks')
map.add_data(data=area_petr.copy(), name='territory_petr')
map.add_data(data=area_vas.copy(), name='territory_vas')
map.add_data(data=border.copy(), name='border')
map.add_data(data = city.copy(), name = 'city')

map
```
